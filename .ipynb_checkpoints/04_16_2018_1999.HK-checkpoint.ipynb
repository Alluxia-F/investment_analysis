```python
import pandas as pd
from pandas_datareader.data import DataReader
from datetime import date,datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_source='yahoo'
drop_col=['Open','High','Low','Close','Adj Close']

start_date=date(2015,1,1)
ticker='1999.HK'

hk_1999=DataReader(ticker,data_source,start_date)
hk_1999['mean']=(hk_1999['Open']+hk_1999['High']+hk_1999['Low']+hk_1999['Close'])/4
hk_1999_change=hk_1999.drop(drop_col,axis=1)
hk_1999_change['pct_change']=hk_1999_change['mean'].pct_change()
     
```

```python
print(hk_1999_change.tail(50))
```

```python
plt.figure(figsize=(60,15))
plt.plot(hk_1999_change['pct_change'],linewidth=8)
plt.title('Man Wah Holdings daily point change',fontsize=40)
plt.xlabel('Date',fontsize=34)
plt.ylabel('PCT Change',fontsize=34)
plt.xticks(fontsize=34,rotation=90)
plt.yticks(fontsize=34)
plt.axhline(0, color='purple',linewidth=6,ls='--')
plt.axvline('2017-6-15',color='red',linewidth=10,label='MuddyWaters')
plt.show()
```
