```python
import pandas as pd
from pandas_datareader.data import DataReader
from datetime import date,datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_source='yahoo'
drop_col=['Open','High','Low','Close','Adj Close']

start_date=date(2017,1,1)
ticker='IQE.L'

iqe=DataReader(ticker,data_source,start_date)
iqe['mean']=(iqe['Open']+iqe['High']+iqe['Low']+iqe['Close'])/4
iqe_change=iqe.drop(drop_col,axis=1)
iqe_change['pct_change']=iqe_change['mean'].pct_change()
     
```

```python
print(iqe_change.tail(50))
```

```python
plt.figure(figsize=(60,15))
plt.plot(iqe_change['pct_change'],linewidth=8)
plt.title('IQE daily point change',fontsize=40)
plt.xlabel('Date',fontsize=34)
plt.ylabel('PCT Change',fontsize=34)
plt.xticks(fontsize=34,rotation=90)
plt.yticks(fontsize=34)
plt.axhline(0, color='purple',linewidth=6,ls='--')
plt.axvline('2018-2-8',color='red',linewidth=10,label='MuddyWaters')
plt.show()
```
