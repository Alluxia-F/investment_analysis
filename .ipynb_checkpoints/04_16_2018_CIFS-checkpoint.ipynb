```python
import pandas as pd
from pandas_datareader.data import DataReader
from datetime import date,datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_source='yahoo'
drop_col=['Open','High','Low','Close','Adj Close']

start_date=date(2017,8,1)
ticker='CIFS'

cifs=DataReader(ticker,data_source,start_date)
cifs['mean']=(cifs['Open']+cifs['High']+cifs['Low']+cifs['Close'])/4
cifs_change=cifs.drop(drop_col,axis=1)
cifs_change['pct_change']=cifs_change['mean'].pct_change()
    
    
```

```python
print(cifs_change.tail(50))
```

```python
import matplotlib.ticker as plticker

loc = plticker.MultipleLocator(base=1.0)
plt.figure(figsize=(60,15))
plt.plot(cifs_change['pct_change'],linewidth=8)
plt.title('CIFS daily point change',fontsize=40)
plt.xlabel('Date',fontsize=34)
plt.ylabel('PCT Change',fontsize=34)
plt.xticks(fontsize=34,rotation=90)
plt.yticks(fontsize=34)
plt.axhline(0, color='purple',linewidth=6,ls='--')
plt.axvline('2017-12-20',linewidth=8,color='red',label='MuddyWaters Shorts')
plt.show()
```
